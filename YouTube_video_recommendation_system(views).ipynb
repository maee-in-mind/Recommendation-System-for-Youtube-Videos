```python
videomat = data.pivot_table(index ='video_id',columns ='title', values ='category_id')
```

```python
videomat.head()
```

```python
ToOurDaughter_cat= videomat['To Our Daughter']
```

```python
ToOurDaughter_cat.head()
```

```python
SimilarTo_ToOurDaughter_cat=videomat.corrwith(ToOurDaughter_cat)
```

```python
corr_ToOurDaughter_cat= pd.DataFrame(SimilarTo_ToOurDaughter_cat, columns =['Correlation']) 
```
